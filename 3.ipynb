```python
train_x = train_x / 255.0
test_x = test_x / 255.0
```

```python
model = Sequential()
model.add(Conv2D(5, (3, 3), activation='sigmoid', kernel_initializer='he_uniform', padding='same', input_shape=(32, 32, 3)))
model.add(Conv2D(5, (3, 3), activation='sigmoid', kernel_initializer='he_uniform', padding='same', input_shape=(32, 32, 3)))
model.add(MaxPooling2D((8, 8)))
model.add(Flatten())
model.add(Dense(10, activation='softmax'))
opt = SGD(lr=0.001, momentum=0.9)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
def summarize_diagnostics(history):
    g1 = pyplot.subplot(211)
    pyplot.title('Loss')
    g1.set_ylim(0, 2.5)
    g1.plot(history.history['loss'], color='blue', label='train')
    g1.plot(history.history['val_loss'], color='orange', label='test')
    # plot accuracy
    g2 = pyplot.subplot(212)
    pyplot.title('Accuracy')
    g2.set_ylim(0, 1)
    g2.plot(history.history['accuracy'], color='blue', label='train')
    g2.plot(history.history['val_accuracy'], color='orange', label='test')
```

```python
summarize_diagnostics(history)
```

```python
def add_block_of_layers(model, filter_num, activation):
    model.add(Conv2D(filter_num, (3, 3), activation=activation, kernel_initializer='he_uniform', padding='same', input_shape=(32, 32, 3)))
    model.add(Conv2D(filter_num, (3, 3), activation=activation, kernel_initializer='he_uniform', padding='same'))
    model.add(MaxPooling2D((2, 2)))
```

```python
from tensorflow.keras.layers import BatchNormalization

def add_block_of_layers_with_normalization(model, filter_num, activation):
    model.add(Conv2D(filter_num, (3, 3), activation=activation, kernel_initializer='he_uniform', padding='same', input_shape=(32, 32, 3)))
    model.add(BatchNormalization())
    model.add(Conv2D(filter_num, (3, 3), activation=activation, kernel_initializer='he_uniform', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D((2, 2)))
```

```python
model_6 = Sequential()
add_block_of_layers_with_normalization(model_6, 20, 'relu')
add_block_of_layers_with_normalization(model_6, 40, 'relu')
add_block_of_layers_with_normalization(model_6, 80, 'relu')
add_block_of_layers_with_normalization(model_6, 160, 'relu')
model_6.add(Flatten())
model_6.add(Dense(10, activation='softmax'))
# opt = SGD(lr=0.001, momentum=0.9)
model_6.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
history_6 = model_6.fit(train_x, train_y, epochs=30, batch_size=64, validation_data=(test_x, test_y), verbose=1)
```

```python
summarize_diagnostics(history_6)
```

```python
from tensorflow.keras.layers import Dropout

def add_block_of_layers_with_normalization_and_dropout(model, filter_num, activation, rate):
    model.add(Conv2D(filter_num, (3, 3), activation=activation, kernel_initializer='he_uniform', padding='same', input_shape=(32, 32, 3)))
    model.add(BatchNormalization())
    model.add(Conv2D(filter_num, (3, 3), activation=activation, kernel_initializer='he_uniform', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(rate = rate))
```

```python
model_7 = Sequential()
add_block_of_layers_with_normalization_and_dropout(model_7, 20, 'relu', 0.1)
add_block_of_layers_with_normalization_and_dropout(model_7, 40, 'relu', 0.1)
add_block_of_layers_with_normalization_and_dropout(model_7, 80, 'relu', 0.1)
add_block_of_layers_with_normalization_and_dropout(model_7, 160, 'relu', 0.1)
model_7.add(Flatten())
model_7.add(Dense(10, activation='softmax'))
# opt = SGD(lr=0.001, momentum=0.9)
model_7.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
history_7 = model_7.fit(train_x, train_y, epochs=30, batch_size=64, validation_data=(test_x, test_y), verbose=1)
```

```python
summarize_diagnostics(history_7)
```
